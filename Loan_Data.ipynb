```python
import pandas as pd

# Dataset load karo
df = pd.read_csv('loan_data.csv')

# Basic info
print("Shape:", df.shape)
print("\nColumn names:\n", df.columns.tolist())
print("\nData types:\n", df.dtypes)
print("\nFirst 5 rows:")
df.head()
```

```python
# Missing values check
print("Missing values:\n", df.isnull().sum())

# Duplicate rows check
print("\nDuplicate rows:", df.duplicated().sum())

# Summary statistics
df.describe()
```

```python
# Categorical column check
print("Unique values in 'purpose':\n", df['purpose'].unique())
print("\nCount of each purpose:\n", df['purpose'].value_counts())
```

```python
# Column names clean karo — dots ko underscore me convert karo
df.columns = df.columns.str.replace('.', '_', regex=False)

print("New column names:\n", df.columns.tolist())
```

```python
# Outlier check karne ke liye key numeric columns ka boxplot-style summary
import numpy as np

numeric_cols = ['int_rate', 'installment', 'log_annual_inc', 'dti', 'fico',
                 'days_with_cr_line', 'revol_bal', 'revol_util']

for col in numeric_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"{col}: {len(outliers)} outliers")
```

```python
# 'purpose' column ko category type me convert karo (memory efficient + clean)
df['purpose'] = df['purpose'].astype('category')

print(df['purpose'].cat.categories)
print("\nUpdated dtypes:\n", df.dtypes)
```

```python
# Sanity check - koi impossible/negative values to nahi hain
print("Negative values check:")
for col in ['int_rate', 'installment', 'fico', 'dti', 'revol_bal']:
    neg_count = (df[col] < 0).sum()
    print(f"{col}: {neg_count} negative values")

# FICO score range check (valid range 300-850 hota hai)
print("\nFICO range:", df['fico'].min(), "-", df['fico'].max())
```

```python
# Extreme outlier flag column bana rahe hain (delete nahi kar rahe, sirf flag)
df['is_outlier'] = 0

for col in ['revol_bal', 'days_with_cr_line', 'dti']:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 3 * iqr   # 3*IQR = extreme outliers only
    upper = q3 + 3 * iqr
    df.loc[(df[col] < lower) | (df[col] > upper), 'is_outlier'] = 1

print("Total flagged extreme outliers:", df['is_outlier'].sum())
```

```python
# Final check before export
print("Final shape:", df.shape)
print("\nFinal dtypes:\n", df.dtypes)
print("\nAny nulls left:", df.isnull().sum().sum())

# Cleaned dataset ko CSV me export karo
df.to_csv('loan_data_cleaned.csv', index=False)
print("\n✅ File saved as loan_data_cleaned.csv")

# Colab se download karne ke liye
from google.colab import files
files.download('loan_data_cleaned.csv')
```
